# KMeans + RandomForest 기반 음식 소비 패턴 예측 모델

# 클러스터링 + 분류 모델(RandomForest) 을 연계한 복합 모델

```python
# ✅ 클러스터링 기반 음식 그룹 예측 모델 (로컬 실행용 전체 코드)

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 데이터 불러오기
df = pd.read_csv('/content/drive/MyDrive/ai/alone/merge_trends_weather_long_ml.csv')  # 파일 경로 수정 필요

# 2. 날짜 전처리 및 결측치 처리
df['Date'] = pd.to_datetime(df['Date'])
df['Date'] = df['Date'].dt.strftime('%m-%d')
df['Precipitation'] = df['Precipitation'].replace(-9.0, np.nan).fillna(0)

# 3. 라벨 인코딩 (GroupName은 이번엔 예측 타깃이 아님)
label_cols = ['Gender', 'Age_Group']
for col in label_cols:
    le = LabelEncoder()
    df[col + '_Label'] = le.fit_transform(df[col])

# 4. 클러스터링에 사용할 피처 선택
features = ['Search_Count', 'Temperature', 'Humidity', 'Precipitation', 'Month', 'Weekday', 'Gender_Label', 'Age_Group_Label']
X_cluster = df[features]

# 5. 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# 6. 클러스터링 (KMeans)
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['Cluster_Label'] = kmeans.fit_predict(X_scaled)

# 7. 클러스터 예측용 데이터 준비
X = X_cluster
y = df['Cluster_Label']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# 8. 랜덤포레스트 학습
clf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
clf.fit(X_train, y_train)

# 9. 평가
y_pred = clf.predict(X_test)
print("\n🎯 정확도:", accuracy_score(y_test, y_pred))
print("\n📋 분류 리포트:\n", classification_report(y_test, y_pred))

# 10. 혼동 행렬
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Cluster')
plt.ylabel('Actual Cluster')
plt.show()

# 11. 피처 중요도
importances = clf.feature_importances_
feature_names = X.columns
feat_imp_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feat_imp_df)
plt.title('📌 Feature Importance for Cluster Prediction')
plt.tight_layout()
plt.show()

# 12. 클러스터별 통계 확인
print("\n📊 클러스터별 평균 특성:")
print(df.groupby('Cluster_Label')[features].mean())

# 13. 클러스터별 대표 음식 (GroupName 빈도 기준)
print("\n🍽 클러스터별 대표 음식:")
if 'GroupName' in df.columns:
    for i in range(n_clusters):
        top = df[df['Cluster_Label'] == i]['GroupName'].value_counts().idxmax()
        print(f"클러스터 {i} 대표 음식: {top}")
```

```python
# ✅ 클러스터링 기반 음식 그룹 예측 모델 (Search_Count 클러스터링 전용, 예측 제외)

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 데이터 불러오기
df = pd.read_csv('/content/drive/MyDrive/ai/alone/merge_trends_weather_long_ml.csv')  # 파일 경로 수정 필요

# 2. 날짜 전처리 및 결측치 처리
df['Date'] = pd.to_datetime(df['Date'])
df['Date'] = df['Date'].dt.strftime('%m-%d')
df['Precipitation'] = df['Precipitation'].replace(-9.0, np.nan).fillna(0)

# 3. 라벨 인코딩 (GroupName은 이번엔 예측 타깃이 아님)
label_cols = ['Gender', 'Age_Group']
for col in label_cols:
    le = LabelEncoder()
    df[col + '_Label'] = le.fit_transform(df[col])

# 4. 클러스터링용 피처 (Search_Count 포함)
features_for_cluster = ['Search_Count', 'Temperature', 'Humidity', 'Precipitation', 'Month', 'Weekday', 'Gender_Label', 'Age_Group_Label']
X_cluster = df[features_for_cluster]

# 5. 예측용 피처 (Search_Count 제외)
features_for_predict = ['Temperature', 'Humidity', 'Precipitation', 'Month', 'Weekday', 'Gender_Label', 'Age_Group_Label']
X_predict = df[features_for_predict]

# 6. 표준화 (클러스터링용)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# 7. Elbow method로 최적 클러스터 수 찾기 (시각화)
inertias = []
k_range = range(2, 10)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertias, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# 8. 클러스터링 수행 (Search_Count 기반)
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['Cluster_Label'] = kmeans.fit_predict(X_scaled)

# 9. 예측 모델 학습 (Search_Count 제외)
X = X_predict
y = df['Cluster_Label']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

clf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
clf.fit(X_train, y_train)

# 10. 평가
y_pred = clf.predict(X_test)
print("\n🎯 정확도:", accuracy_score(y_test, y_pred))
print("\n📋 분류 리포트:\n", classification_report(y_test, y_pred))

# 11. 혼동 행렬
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Cluster')
plt.ylabel('Actual Cluster')
plt.show()

# 12. 피처 중요도
importances = clf.feature_importances_
feature_names = X.columns
feat_imp_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feat_imp_df)
plt.title('📌 Feature Importance for Cluster Prediction')
plt.tight_layout()
plt.show()

# 13. 클러스터별 평균 특성
print("\n📊 클러스터별 평균 특성:")
print(df.groupby('Cluster_Label')[features_for_predict].mean())

# 14. 클러스터별 대표 음식 시각화
if 'GroupName' in df.columns:
    cluster_group = df.groupby(['Cluster_Label', 'GroupName']).size().reset_index(name='Count')
    plt.figure(figsize=(12, 6))
    sns.barplot(data=cluster_group, x='Cluster_Label', y='Count', hue='GroupName')
    plt.title('🍽 클러스터별 음식 그룹 분포')
    plt.xlabel('Cluster')
    plt.ylabel('Count')
    plt.legend(title='GroupName', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# 15. 예측 결과 저장
results_df = X_test.copy()
results_df['Actual_Cluster'] = y_test
results_df['Predicted_Cluster'] = y_pred
results_df.to_csv('cluster_prediction_results.csv', index=False)
print("\n📁 예측 결과가 'cluster_prediction_results.csv'로 저장되었습니다.")

# 16. 사용자 입력 예측 함수 (Search_Count 제외)
def predict_user_cluster(input_dict):
    input_df = pd.DataFrame([input_dict])
    predicted_cluster = clf.predict(input_df)[0]
    print(f"\n🧑 입력한 조건은 클러스터 {predicted_cluster}에 속합니다.")
    return predicted_cluster

# 17. 사용자 예시 입력 (Search_Count 없음)
user_input = {
    'Temperature': 12.5,
    'Humidity': 60.0,
    'Precipitation': 0.0,
    'Month': 3,
    'Weekday': 5,
    'Gender_Label': 1,  # 여성
    'Age_Group_Label': 0  # 청년층
}
predict_user_cluster(user_input)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 데이터 불러오기
df = pd.read_csv('/content/drive/MyDrive/ai/alone/merge_trends_weather_long_ml.csv')  # 경로 확인

# 2. 날짜 전처리 및 결측치 처리
df['Date'] = pd.to_datetime(df['Date'])
df['Date'] = df['Date'].dt.strftime('%m-%d')
df['Precipitation'] = df['Precipitation'].replace(-9.0, np.nan).fillna(0)

# 3. 라벨 인코딩
label_cols = ['Gender', 'Age_Group']
for col in label_cols:
    le = LabelEncoder()
    df[col + '_Label'] = le.fit_transform(df[col])

# 4. 로그 변환된 Search_Count 생성
df['Log_Search_Count'] = np.log1p(df['Search_Count'])

# 5. 클러스터링용 피처
features_for_cluster = ['Log_Search_Count', 'Temperature', 'Humidity', 'Precipitation',
                        'Month', 'Weekday', 'Gender_Label', 'Age_Group_Label']
X_cluster = df[features_for_cluster]

# 6. 예측용 피처 (Search_Count 제외)
features_for_predict = ['Temperature', 'Humidity', 'Precipitation', 'Month',
                        'Weekday', 'Gender_Label', 'Age_Group_Label']
X_predict = df[features_for_predict]

# 7. 표준화
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# 8. Elbow Method
inertias = []
k_range = range(2, 10)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertias, marker='o')
plt.title('Elbow Method with Log(Search_Count)')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# 9. 클러스터링 수행 (k = 5)
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['Cluster_Label'] = kmeans.fit_predict(X_scaled)

# ✅ 10. Silhouette Score 평가 (샘플링)
print("✔ Silhouette Score 계산 중... (샘플 사용)")
sample_size = 5000
X_sample = X_scaled[:sample_size]
labels_sample = df['Cluster_Label'][:sample_size]
sil_score = silhouette_score(X_sample, labels_sample)
print(f"\n🔍 Silhouette Score (샘플 {sample_size}개 기준): {sil_score:.4f}")

# 11. 예측 모델 학습
X = X_predict
y = df['Cluster_Label']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

clf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
clf.fit(X_train, y_train)

# 12. 평가
y_pred = clf.predict(X_test)
print("\n🎯 정확도:", accuracy_score(y_test, y_pred))
print("\n📋 분류 리포트:\n", classification_report(y_test, y_pred))

# 13. 혼동 행렬
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Cluster')
plt.ylabel('Actual Cluster')
plt.show()

# 14. 피처 중요도
importances = clf.feature_importances_
feature_names = X.columns
feat_imp_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feat_imp_df)
plt.title('📌 Feature Importance for Cluster Prediction')
plt.tight_layout()
plt.show()

# 15. 클러스터별 평균 특성
print("\n📊 클러스터별 평균 특성:")
print(df.groupby('Cluster_Label')[features_for_predict].mean())

# 16. 클러스터별 대표 음식 시각화
if 'GroupName' in df.columns:
    cluster_group = df.groupby(['Cluster_Label', 'GroupName']).size().reset_index(name='Count')
    plt.figure(figsize=(12, 6))
    sns.barplot(data=cluster_group, x='Cluster_Label', y='Count', hue='GroupName')
    plt.title('🍽 클러스터별 음식 그룹 분포')
    plt.xlabel('Cluster')
    plt.ylabel('Count')
    plt.legend(title='GroupName', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# 17. 예측 결과 저장
results_df = X_test.copy()
results_df['Actual_Cluster'] = y_test
results_df['Predicted_Cluster'] = y_pred
results_df.to_csv('cluster_prediction_results.csv', index=False)
print("\n📁 예측 결과가 'cluster_prediction_results.csv'로 저장되었습니다.")

# 18. 사용자 입력 예측 함수
def predict_user_cluster(input_dict):
    input_df = pd.DataFrame([input_dict])
    predicted_cluster = clf.predict(input_df)[0]
    print(f"\n🧑 입력한 조건은 클러스터 {predicted_cluster}에 속합니다.")
    return predicted_cluster

# 19. 사용자 예시 입력
user_input = {
    'Temperature': 12.5,
    'Humidity': 60.0,
    'Precipitation': 0.0,
    'Month': 3,
    'Weekday': 5,
    'Gender_Label': 1,  # 여성
    'Age_Group_Label': 0  # 청년층
}
predict_user_cluster(user_input)
```
