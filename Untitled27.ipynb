```python
# ‚úÖ ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ Í∏∞Î∞ò ÏùåÏãù Í∑∏Î£π ÏòàÏ∏° Î™®Îç∏ (Î°úÏª¨ Ïã§ÌñâÏö© Ï†ÑÏ≤¥ ÏΩîÎìú)

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
df = pd.read_csv('/content/drive/MyDrive/ai/alone/merge_trends_weather_long_ml.csv')  # ÌååÏùº Í≤ΩÎ°ú ÏàòÏ†ï ÌïÑÏöî

# 2. ÎÇ†Ïßú Ï†ÑÏ≤òÎ¶¨ Î∞è Í≤∞Ï∏°Ïπò Ï≤òÎ¶¨
df['Date'] = pd.to_datetime(df['Date'])
df['Date'] = df['Date'].dt.strftime('%m-%d')
df['Precipitation'] = df['Precipitation'].replace(-9.0, np.nan).fillna(0)

# 3. ÎùºÎ≤® Ïù∏ÏΩîÎî© (GroupNameÏùÄ Ïù¥Î≤àÏóî ÏòàÏ∏° ÌÉÄÍπÉÏù¥ ÏïÑÎãò)
label_cols = ['Gender', 'Age_Group']
for col in label_cols:
    le = LabelEncoder()
    df[col + '_Label'] = le.fit_transform(df[col])

# 4. ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅÏóê ÏÇ¨Ïö©Ìï† ÌîºÏ≤ò ÏÑ†ÌÉù
features = ['Search_Count', 'Temperature', 'Humidity', 'Precipitation', 'Month', 'Weekday', 'Gender_Label', 'Age_Group_Label']
X_cluster = df[features]

# 5. ÌëúÏ§ÄÌôî
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# 6. ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ (KMeans)
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['Cluster_Label'] = kmeans.fit_predict(X_scaled)

# 7. ÌÅ¥Îü¨Ïä§ÌÑ∞ ÏòàÏ∏°Ïö© Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
X = X_cluster
y = df['Cluster_Label']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

# 8. ÎûúÎç§Ìè¨Î†àÏä§Ìä∏ ÌïôÏäµ
clf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
clf.fit(X_train, y_train)

# 9. ÌèâÍ∞Ä
y_pred = clf.predict(X_test)
print("\nüéØ Ï†ïÌôïÎèÑ:", accuracy_score(y_test, y_pred))
print("\nüìã Î∂ÑÎ•ò Î¶¨Ìè¨Ìä∏:\n", classification_report(y_test, y_pred))

# 10. ÌòºÎèô ÌñâÎ†¨
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Cluster')
plt.ylabel('Actual Cluster')
plt.show()

# 11. ÌîºÏ≤ò Ï§ëÏöîÎèÑ
importances = clf.feature_importances_
feature_names = X.columns
feat_imp_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feat_imp_df)
plt.title('üìå Feature Importance for Cluster Prediction')
plt.tight_layout()
plt.show()

# 12. ÌÅ¥Îü¨Ïä§ÌÑ∞Î≥Ñ ÌÜµÍ≥Ñ ÌôïÏù∏
print("\nüìä ÌÅ¥Îü¨Ïä§ÌÑ∞Î≥Ñ ÌèâÍ∑† ÌäπÏÑ±:")
print(df.groupby('Cluster_Label')[features].mean())

# 13. ÌÅ¥Îü¨Ïä§ÌÑ∞Î≥Ñ ÎåÄÌëú ÏùåÏãù (GroupName ÎπàÎèÑ Í∏∞Ï§Ä)
print("\nüçΩ ÌÅ¥Îü¨Ïä§ÌÑ∞Î≥Ñ ÎåÄÌëú ÏùåÏãù:")
if 'GroupName' in df.columns:
    for i in range(n_clusters):
        top = df[df['Cluster_Label'] == i]['GroupName'].value_counts().idxmax()
        print(f"ÌÅ¥Îü¨Ïä§ÌÑ∞ {i} ÎåÄÌëú ÏùåÏãù: {top}")
```

```python
# ‚úÖ ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ Í∏∞Î∞ò ÏùåÏãù Í∑∏Î£π ÏòàÏ∏° Î™®Îç∏ (Search_Count ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ Ï†ÑÏö©, ÏòàÏ∏° Ï†úÏô∏)

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Îç∞Ïù¥ÌÑ∞ Î∂àÎü¨Ïò§Í∏∞
df = pd.read_csv('/content/drive/MyDrive/ai/alone/merge_trends_weather_long_ml.csv')  # ÌååÏùº Í≤ΩÎ°ú ÏàòÏ†ï ÌïÑÏöî

# 2. ÎÇ†Ïßú Ï†ÑÏ≤òÎ¶¨ Î∞è Í≤∞Ï∏°Ïπò Ï≤òÎ¶¨
df['Date'] = pd.to_datetime(df['Date'])
df['Date'] = df['Date'].dt.strftime('%m-%d')
df['Precipitation'] = df['Precipitation'].replace(-9.0, np.nan).fillna(0)

# 3. ÎùºÎ≤® Ïù∏ÏΩîÎî© (GroupNameÏùÄ Ïù¥Î≤àÏóî ÏòàÏ∏° ÌÉÄÍπÉÏù¥ ÏïÑÎãò)
label_cols = ['Gender', 'Age_Group']
for col in label_cols:
    le = LabelEncoder()
    df[col + '_Label'] = le.fit_transform(df[col])

# 4. ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅÏö© ÌîºÏ≤ò (Search_Count Ìè¨Ìï®)
features_for_cluster = ['Search_Count', 'Temperature', 'Humidity', 'Precipitation', 'Month', 'Weekday', 'Gender_Label', 'Age_Group_Label']
X_cluster = df[features_for_cluster]

# 5. ÏòàÏ∏°Ïö© ÌîºÏ≤ò (Search_Count Ï†úÏô∏)
features_for_predict = ['Temperature', 'Humidity', 'Precipitation', 'Month', 'Weekday', 'Gender_Label', 'Age_Group_Label']
X_predict = df[features_for_predict]

# 6. ÌëúÏ§ÄÌôî (ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅÏö©)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_cluster)

# 7. Elbow methodÎ°ú ÏµúÏ†Å ÌÅ¥Îü¨Ïä§ÌÑ∞ Ïàò Ï∞æÍ∏∞ (ÏãúÍ∞ÅÌôî)
inertias = []
k_range = range(2, 10)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(k_range, inertias, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()

# 8. ÌÅ¥Îü¨Ïä§ÌÑ∞ÎßÅ ÏàòÌñâ (Search_Count Í∏∞Î∞ò)
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
df['Cluster_Label'] = kmeans.fit_predict(X_scaled)

# 9. ÏòàÏ∏° Î™®Îç∏ ÌïôÏäµ (Search_Count Ï†úÏô∏)
X = X_predict
y = df['Cluster_Label']
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

clf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
clf.fit(X_train, y_train)

# 10. ÌèâÍ∞Ä
y_pred = clf.predict(X_test)
print("\nüéØ Ï†ïÌôïÎèÑ:", accuracy_score(y_test, y_pred))
print("\nüìã Î∂ÑÎ•ò Î¶¨Ìè¨Ìä∏:\n", classification_report(y_test, y_pred))

# 11. ÌòºÎèô ÌñâÎ†¨
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted Cluster')
plt.ylabel('Actual Cluster')
plt.show()

# 12. ÌîºÏ≤ò Ï§ëÏöîÎèÑ
importances = clf.feature_importances_
feature_names = X.columns
feat_imp_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feat_imp_df)
plt.title('üìå Feature Importance for Cluster Prediction')
plt.tight_layout()
plt.show()

# 13. ÌÅ¥Îü¨Ïä§ÌÑ∞Î≥Ñ ÌèâÍ∑† ÌäπÏÑ±
print("\nüìä ÌÅ¥Îü¨Ïä§ÌÑ∞Î≥Ñ ÌèâÍ∑† ÌäπÏÑ±:")
print(df.groupby('Cluster_Label')[features_for_predict].mean())

# 14. ÌÅ¥Îü¨Ïä§ÌÑ∞Î≥Ñ ÎåÄÌëú ÏùåÏãù ÏãúÍ∞ÅÌôî
if 'GroupName' in df.columns:
    cluster_group = df.groupby(['Cluster_Label', 'GroupName']).size().reset_index(name='Count')
    plt.figure(figsize=(12, 6))
    sns.barplot(data=cluster_group, x='Cluster_Label', y='Count', hue='GroupName')
    plt.title('üçΩ ÌÅ¥Îü¨Ïä§ÌÑ∞Î≥Ñ ÏùåÏãù Í∑∏Î£π Î∂ÑÌè¨')
    plt.xlabel('Cluster')
    plt.ylabel('Count')
    plt.legend(title='GroupName', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()

# 15. ÏòàÏ∏° Í≤∞Í≥º Ï†ÄÏû•
results_df = X_test.copy()
results_df['Actual_Cluster'] = y_test
results_df['Predicted_Cluster'] = y_pred
results_df.to_csv('cluster_prediction_results.csv', index=False)
print("\nüìÅ ÏòàÏ∏° Í≤∞Í≥ºÍ∞Ä 'cluster_prediction_results.csv'Î°ú Ï†ÄÏû•ÎêòÏóàÏäµÎãàÎã§.")

# 16. ÏÇ¨Ïö©Ïûê ÏûÖÎ†• ÏòàÏ∏° Ìï®Ïàò (Search_Count Ï†úÏô∏)
def predict_user_cluster(input_dict):
    input_df = pd.DataFrame([input_dict])
    predicted_cluster = clf.predict(input_df)[0]
    print(f"\nüßë ÏûÖÎ†•Ìïú Ï°∞Í±¥ÏùÄ ÌÅ¥Îü¨Ïä§ÌÑ∞ {predicted_cluster}Ïóê ÏÜçÌï©ÎãàÎã§.")
    return predicted_cluster

# 17. ÏÇ¨Ïö©Ïûê ÏòàÏãú ÏûÖÎ†• (Search_Count ÏóÜÏùå)
user_input = {
    'Temperature': 12.5,
    'Humidity': 60.0,
    'Precipitation': 0.0,
    'Month': 3,
    'Weekday': 5,
    'Gender_Label': 1,  # Ïó¨ÏÑ±
    'Age_Group_Label': 0  # Ï≤≠ÎÖÑÏ∏µ
}
predict_user_cluster(user_input)
```
