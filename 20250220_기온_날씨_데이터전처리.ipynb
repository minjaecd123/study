# 음식별 가장많이 검색되는 날짜 데이터프레임에 저장 (네이버 트렌드 활용)

```python
import requests
import json
import pandas as pd
import time

# 네이버 API 정보
CLIENT_ID = "JNIJiajEDYjf8csQs947"
CLIENT_SECRET = "70dBGBsyyR"

url = "https://openapi.naver.com/v1/datalab/search"

# 검색할 기간 설정
START_DATE = "2022-01-01"
END_DATE = "2024-12-31"
TIME_UNIT = "month"

# 데이터프레임 예제 (실제 데이터로 대체 가능)
df = df[["대표식품명", "식품대분류명"]]

# 검색 데이터를 저장할 리스트
search_results = []

# 각 대표식품명을 하나씩 검색하여 데이터 수집
for food_name in df["대표식품명"]:
    params = {
        "startDate": START_DATE,
        "endDate": END_DATE,
        "timeUnit": TIME_UNIT,
        "keywordGroups": [{"groupName": food_name, "keywords": [food_name]}]
    }

    headers = {
        "X-Naver-Client-Id": CLIENT_ID,
        "X-Naver-Client-Secret": CLIENT_SECRET,
        "Content-Type": "application/json"
    }

    response = requests.post(url, data=json.dumps(params), headers=headers)

    if response.status_code == 200:
        data = response.json()

        # 검색량이 가장 높은 기간 찾기
        max_ratio = 0
        max_period = ""

        for entry in data['results'][0]['data']:
            if entry['ratio'] > max_ratio:
                max_ratio = entry['ratio']
                max_period = entry['period']

        search_results.append({"대표식품명": food_name, "최고검색기간": max_period, "최고검색량": max_ratio})

    else:
        print(f"API 요청 실패: {food_name}, 상태 코드: {response.status_code}")
        search_results.append({"대표식품명": food_name, "최고검색기간": "실패", "최고검색량": 0})

    # API 요청 간격 조정 (네이버 API 제한 방지)
    time.sleep(1)

# 결과를 새로운 데이터프레임으로 변환
search_df = pd.DataFrame(search_results)

# 기존 데이터프레임과 병합
result_df = df.merge(search_df, on="대표식품명")

# 결과를 CSV 파일로 저장
file_path = "/content/drive/MyDrive/사계밥상"
result_df.to_csv(file_path, index=False, encoding="utf-8-sig")

# 사용자에게 파일 다운로드 링크 제공
print(f"파일이 저장되었습니다: {file_path}")
```
