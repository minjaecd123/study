#분류작업중 (미완성)

```python
import pandas as pd
import numpy as np

# 1) 분류하고자 하는 음식 이름 리스트(일부 예시만 사용)
food_list = df['대표식품명']

# 2) Pandas DataFrame 생성
data = pd.DataFrame(food_list, columns=['대표식품명'])

# 3) 카테고리별 키워드 사전 정의(예시)
categories = {
    '한식': [
        "김치", "찌개", "비빔밥", "된장찌개", "불고기", "감자탕", "나물", "전", "떡", "튀김", "구이", "고구마", "게장",'겉절이','겨자채',"볶음","고춧가루","장류"
        "생채", "무침", "회", "국", "탕", "김밥", "국수", "만두", "삼겹살", "갈비","밥",'채','찜',"기름장","김","죽",
        "죽", "밥", "잡채", "장아찌", "조림", "고추장", "찜", "갈비찜", "비빔국수", "수육", "보쌈", "전", "부침",
        "고등어구이", "황태국밥", "삼계탕", "갈비탕", "순두부찌개", "제육볶음", "닭갈비", "칼국수", "누룽지탕", "순대국"
    ],
    '중식': [
        "짜장면", "짬뽕", "탕수육", "마파두부", "깐풍기", "멘보샤", "유산슬", "볶음밥", "양장피",
        "훠궈", "팔보채", "오향장육", "마라탕", "고추잡채", "계란탕",
        "춘권", "딤섬", "꿔바로우", "마라샹궈","기스면","깐풍"
    ],
    '일식': [
        "라멘", "스시", "초밥", "가츠동", "덴푸라", "우동", "미소된장국", "사시미", "돈부리", "규동",
        "오코노미야키", "야키소바", "규카츠",'나베',
        "가라아게", "타코야키", "오니기리", "카레우동"
    ],
    '양식': [
        "피자", "파스타", "스테이크", "햄버거", "샌드위치", "핫도그", "스프", "카레라이스", "샐러드",
        "리조또", "그라탕", "퀘사디아", "타코", "파니니", "스파게티",
        "라자냐", "뇨끼", "포카치아", "미트로프"
    ],
    '간식': [
        "빵", "쿠키", "케이크", "커피", "아이스크림", "호떡", "과자", "차", "떡볶이", "라떼", "핫케이크",
        "스콘", "머핀", "도넛", "에그타르트", "빙수", "과일", "주스",
        "팝콘", "마카롱", "와플", "초콜릿","견과","경단","계란빵","효모빵"
    ]
}



# 4) 분류 함수 정의
def classify_dish(dish_name):
    # 각 카테고리를 순회하며 dish_name에 키워드가 포함되어 있는지 검사
    for category, keywords in categories.items():
        for keyword in keywords:
            # 음식명 안에 해당 keyword가 있으면 해당 카테고리로 분류
            if keyword in dish_name:
                return category
    # 어떤 카테고리에도 속하지 않으면 '기타'로 분류
    return '기타'

# 5) 분류 적용
df['category'] = df['대표식품명'].apply(classify_dish)
df['category'] = np.where(df['식품대분류명'] == '음료 및 차류',"간식",
                 np.where(df['식품대분류명'] == '김치류',"한식",
                 np.where(df['식품대분류명'] == '김치류',"한식",
                 np.where(df['식품대분류명'] == '전·적 및 부침류',"한식",df['대표식품명'].apply(classify_dish)
))))
# 결과 확인
df['category'].value_counts() # 아직 미분류(기타)가 많다

#df['대표식품명'].apply(classify_dish)
```
